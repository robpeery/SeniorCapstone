```python
standf = pstpitchdata[['Name', 'Age', '#days', 'Lev', 'Tm', 'G', 'GS', 'W', 'L', 'SV', 'IP',
       'H', 'R', 'ER', 'BB', 'SO', 'HR', 'HBP', 'ERA', 'AB', '2B', '3B', 'IBB',
       'GDP', 'SF', 'SB', 'CS', 'PO', 'BF', 'Pit', 'Str', 'StL', 'StS',
       'GB/FB', 'LD', 'PU', 'WHIP', 'BAbip', 'SO9', 'SO/W']]
prestandf = prepitchdata[['Name', 'Age', '#days', 'Lev', 'Tm', 'G', 'GS', 'W', 'L', 'SV', 'IP',
       'H', 'R', 'ER', 'BB', 'SO', 'HR', 'HBP', 'ERA', 'AB', '2B', '3B', 'IBB',
       'GDP', 'SF', 'SB', 'CS', 'PO', 'BF', 'Pit', 'Str', 'StL', 'StS',
       'GB/FB', 'LD', 'PU', 'WHIP', 'BAbip', 'SO9', 'SO/W']]
#Put standard data into dataframes
```

```python
#Rename columns after June 20th to pst_stat
pstdf = standf.rename(columns={'Age':'pst_Age', '#days':'pst_days', 'Lev':'pst_lev', 
                               'Tm':'pst_team', 'G':'pst_G', 'GS':'pst_GS', 'W':'pst_W', 'L':'pst_L', 'SV':'pst_SV', 
                               'IP':'pst_IP',
       'H':'pst_H', 'R':'pst_R', 'ER':'pst_ER', 'BB':'pst_BB', 'SO':'pst_SO', 'HR':'pst_HR', 'HBP':'pst_HBP',
                               'ERA':'pst_ERA', 'AB':'pst_AB', '2B':'pst_2B', '3B':'pst_3B', 'IBB':'pst_IBB',
       'GDP':'pst_GDP','SF':'pst_SF', 'SB':'pst_SB', 'CS':'pst_CS', 'PO':'pst_PO', 'BF':'pst_BF', 'Pit':'pst_Pit',
                               'Str':'pst_Str', 'StL':'pst_StL', 'StS':'pst_StS',
       'GB/FB':'pst_GB/FB', 'LD':'pst_LD', 'PU':'pst_PU', 'WHIP':'pst_WHIP', 'BAbip':'pst_BAbip', 'SO9':'pst_SO9', 'SO/W':'pst_SO/W'})
```

```python
standardstatsdf = pstdf.merge(prestandf)
#Merge standard stats
```

```python
standardstatsdf = standardstatsdf.set_index('Name')
#Set index to player name
```

```python
df['release_spin_rate']=df['release_spin_rate'].astype(float)
prebandf['release_spin_rate']=prebandf['release_spin_rate'].astype(float)
#Set spinrate to float for pivot table purposes
```

```python
df3=df.pivot_table(index=['player_name'],values = 'release_spin_rate',columns="pitch_type",aggfunc='mean')
df5=prebandf.pivot_table(index=['player_name'],values = 'release_spin_rate',columns="pitch_type",aggfunc='mean')
# Get average spin rates per pitcher
```

```python
df3 = df3[['FF']]
df5 = df5[['FF']]
#Only fastballs
```

```python
df3 = df3.rename(columns={'FF':'PostB FF'})
df5 = df5.rename(columns={'FF':'PreB FF'})
#Rename to post and pre ban
```

```python
print(df3)
print(df5)
```

```python
spindf = df3.join(df5)
#join post and pre spin
spindf.index
```

```python
spindf = spindf.dropna()
```

```python
namelist = spindf.index
[n.split(", ") for n in namelist]
[n.split(", ")[::-1] for n in namelist]
namelist = [" ".join(n.split(", ")[::-1]) for n in namelist]
#Edit names to max index of Standard Stats
```

```python
spindf.index = namelist
#Assign index to new names
```

```python
df = standardstatsdf.join(spindf)
df
#join indexs
```

```python
df.to_csv('2021StandardPrePost&FastballSpin.csv')
#print to csv
```
