```python
baseball_csv = pd.read_csv('../baseball_csv/post_ban_statcast_pitch.csv')
```

```python
df1= baseball_csv[["game_date","pitcher","strikes","at_bat_number","release_spin_rate"]]
```

```python
df2= df1.dropna()
```

```python
df2
```

```python
df2['release_spin_rate']=df2['release_spin_rate'].astype(float)
#make release spin rate a float (for pivot table purposes)
```

```python
df2
```

```python
df2['%K'] = (df2['strikes']/df2['at_bat_number'])
```

```python
df2
```

```python
df2[['release_spin_rate','%K']].max()
```

```python
df2[['release_spin_rate','%K']].min()
```

```python
df3_mean = df2.groupby('pitcher', as_index=False)['%K','release_spin_rate'].mean()
df3_mean
```

```python
df3_sd = df2.groupby('pitcher', as_index=False)['%K','release_spin_rate'].std()
df3_sd
```

```python
df3_median = df2.groupby('pitcher', as_index=False)['%K','release_spin_rate'].median()
df3_median
```

```python
high_interval = (df3_mean['release_spin_rate']+(2.575*df3_sd['release_spin_rate']/np.sqrt(len(df3_mean))))
low_interval = (df3_mean['release_spin_rate']-(2.575*df3_sd['release_spin_rate']/np.sqrt(len(df3_mean))))
```

```python
spin_conf_high = np.array(high_interval).flatten()
spin_conf_low = np.array(low_interval).flatten()
spin_difference_conf = (spin_conf_high.astype(float) - spin_conf_low.astype(float)).flatten()
np.amax(spin_difference_conf)
```

```python
spin_conf_high = pd.DataFrame({'spin_high_conf': spin_conf_high})
spin_conf_high
```

```python
spin_conf_low = pd.DataFrame({'spin_low_conf': spin_conf_low})
spin_conf_low
```

```python
spin_difference_conf = pd.DataFrame({'spin_difference_conf': spin_difference_conf})
spin_difference_conf
```

```python
player_id = pd.DataFrame({'Player ID': df3_median['pitcher']})
player_id
```

```python
frames = [player_id, spin_conf_low, spin_conf_high, spin_difference_conf]
spin_rate_conf_df = pd.concat(frames, axis=1)
spin_rate_conf_df
```

```python
high_interval_k = (df3_mean['%K']+(2.575*df3_sd['%K']/np.sqrt(len(df3_mean))))
low_interval_k = (df3_mean['%K']-(2.575*df3_sd['%K']/np.sqrt(len(df3_mean))))
```

```python
k_conf_high = np.array(high_interval_k).flatten()
k_conf_low = np.array(low_interval_k).flatten()
k_difference_conf = (k_conf_high.astype(float) - k_conf_low.astype(float)).flatten()
```

```python
k_conf_high = pd.DataFrame({'%k conf_high': k_conf_high})
k_conf_high
```

```python
k_conf_low = pd.DataFrame({'%k conf_low': k_conf_low})
k_conf_low
```

```python
k_difference_conf = pd.DataFrame({'%k difference_conf': k_difference_conf})
k_difference_conf
```

```python
frames = [k_conf_low, k_conf_high, k_difference_conf]
strikeout_rate_df = pd.concat(frames, axis=1)
strikeout_rate_df
```

```python
frames = [spin_rate_conf_df, strikeout_rate_df, df3_mean]
strikeout_rate__spin_rate_df = pd.concat(frames, axis=1)
strikeout_rate__spin_rate_df = strikeout_rate__spin_rate_df.drop(columns=['pitcher'])
strikeout_rate__spin_rate_df = strikeout_rate__spin_rate_df.rename(columns={'%K': '%K mean', 'release_spin_rate': 'spin rate mean'})
strikeout_rate__spin_rate_df
```

```python
import altair as alt
import pandas as pd
import numpy as np

np.random.seed(42)
source = pd.DataFrame(np.cumsum(np.random.randn(100, 3), 0).round(2),
                    columns=['A', 'B', 'C'], index=pd.RangeIndex(100, name='x'))
source = source.reset_index().melt('x', var_name='category', value_name='y')

# Create a selection that chooses the nearest point & selects based on x-value
nearest = alt.selection(type='single', nearest=True, on='mouseover',
                        fields=['x'], empty='none')

# The basic line
line = alt.Chart(source).mark_line(interpolate='basis').encode(
    x='x:Q',
    y='y:Q',
    color='category:N'
)

# Transparent selectors across the chart. This is what tells us
# the x-value of the cursor
selectors = alt.Chart(source).mark_point().encode(
    x='x:Q',
    opacity=alt.value(0),
).add_selection(
    nearest
)

# Draw points on the line, and highlight based on selection
points = line.mark_point().encode(
    opacity=alt.condition(nearest, alt.value(1), alt.value(0))
)

# Draw text labels near the points, and highlight based on selection
text = line.mark_text(align='left', dx=5, dy=-5).encode(
    text=alt.condition(nearest, 'y:Q', alt.value(' '))
)

# Draw a rule at the location of the selection
rules = alt.Chart(source).mark_rule(color='gray').encode(
    x='x:Q',
).transform_filter(
    nearest
)

# Put the five layers into a chart and bind the data
alt.layer(
    line, selectors, points, rules, text
).properties(
    width=600, height=300
)
```
