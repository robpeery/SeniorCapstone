```python
import pandas as pd
import numpy as np
import pybaseball
import matplotlib as plt
from pybaseball import statcast
```

```python
data = statcast(start_dt='2021-03-24', end_dt='2021-10-27')
#get data for specified dates. (Can go back to 2015)
```

```python
data
#print data
```

```python
data.columns
```

```python
df1= data[["game_date","player_name","pitch_type","release_speed","release_spin_rate"]]
#pull relevant columns
```

```python
df2= df1.dropna()
```

```python
df2.dtypes
```

```python
df2
```

```python
df2['release_spin_rate']=df2['release_spin_rate'].astype(float)
#make release spin rate a float (for pivot table purposes)
```

```python
print(df2.dtypes)
```

```python
df2
```

```python
league_average = df2.groupby('game_date', as_index=False)['release_spin_rate'].mean()
league_average
#get league average per day for spin rate
```

```python
df3=df2.pivot_table(index=['game_date'],columns="pitch_type",aggfunc='mean')
#pivot table, average spin rate/release speed for every day since 2015
df3
```

```python
df3.columns
```

```python
idx = pd.IndexSlice
A = df3.loc[:,idx['release_spin_rate','FA']]
B = df3.loc[:,idx['release_spin_rate','CH']]
D = df3.loc[:,idx['release_spin_rate','CU']]
E = df3.loc[:,idx['release_spin_rate','SI']]
F = df3.loc[:,idx['release_spin_rate','FS']]
G = df3.loc[:,idx['release_spin_rate','FC']]
#obtain just spin rate columns
```

```python
df4 = pd.concat([A,B,D,E,F,G], axis=1,join='inner')
#join
```
